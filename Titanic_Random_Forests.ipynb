```python
from sklearn import datasets
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
train_data = pd.read_csv("./Titanic_Dataset/training.csv")
test_data = pd.read_csv("./Titanic_Dataset/testing.csv")
sample_data = pd.read_csv("./Titanic_Dataset/sample.csv")
```

```python
train_data.head()
```

```python
y_train = train_data['Survived']
```

```python
X_train = train_data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
```

```python
from sklearn.ensemble import RandomForestClassifier as rf
```

```python
classifier = rf()
```

```python
classifier.fit(X_train, y_train)
```

```python
pred = classifier.predict(test_data)
```

```python
sample_data['solution'] = pred

sample_data.head()
```

```python
sample_data.drop(['class'], inplace=True, axis = 1)
```

```python
sample_data.head(50)
```

```python
sample_data.to_csv('Random_Forests_Titanic.csv', index=False)
```
